```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial import distance
```

```python
data1 = pd.read_table(r"C:\Users\Dell\Downloads\dataset1.txt",  sep = ' ')
```

```python
data1
```

```python
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
  kmeans = KMeans(n_clusters=k, random_state=0).fit(data1)
  print("For value of K:", k)
  print("Clutering Labels for all data points:")
  labels = kmeans.labels_
  print(kmeans.labels_)
  print("Cluster Points:")
  centroids = kmeans.cluster_centers_
  print(kmeans.cluster_centers_)
  Sum_of_squared_distances.append(kmeans.inertia_)
  plt.scatter(data1.iloc[:, 0] , data1.iloc[:, 1] , c=labels)
  plt.scatter(centroids[:,0] , centroids[:,1] , s = 50, color = 'k')
  plt.legend()
  plt.show()
  print("----------------------------------------------------------------------------------")

```

### Elbow graph

```python
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')
plt.show()
```

### DBSCAN

### Dataset 1    Ques1(b)

```python
class Basic_DBSCAN:
   
    
    def __init__(self, eps, minPts, metric=distance.euclidean):
        self.eps = eps
        self.minPts = minPts
        self.metric = metric
    
    def fit_predict(self, X):
        """
        Parameters:X: An n-dimensional array of numeric vectors to be analyzed
        
        Returns:[n] cluster labels
        """
    
        clusters = [0] * X.shape[0]
        
        simple_DBSCAN(X, clusters, self.eps, self.minPts, self.metric)
        
        return clusters

def simple_DBSCAN(X, clusters, eps, minPts, metric=distance.euclidean):
   
    current = 0
    
    for i in range(0, X.shape[0]):
        if clusters[i] is not 0:
            continue
    
        neighbors = neighborsGen(X, i, eps, metric)

        if len(neighbors) < minPts:
            clusters[i] = -1

        else:
            current += 1
            expand(X, clusters, i, neighbors, current, eps, minPts, metric)
    
    return clusters

def neighborsGen(X, point, eps, metric):
    """
    Generates neighborhood graph for a given point
    """
    
    neighbors = []
    
    for i in range(X.shape[0]):
        if metric(X[point], X[i]) < eps:
            neighbors.append(i)
    
    return neighbors

def expand(X, clusters, point, neighbors, current, eps, minPts, metric):
    """
    Expands cluster from a given point until neighborhood boundaries are reached
    """
    clusters[point] = current
    
    i = 0
    while i < len(neighbors):
        
        nextPoint = neighbors[i]
        
        if clusters[nextPoint] == -1:
            clusters[nextPoint] = current
        
        elif clusters[nextPoint] == 0:
            clusters[nextPoint] = current
            
            nextNeighbors = neighborsGen(X, nextPoint, eps, metric)
            
            if len(nextNeighbors) >= minPts:
                neighbors = neighbors + nextNeighbors
        
        i += 1
```

```python
scanner = Basic_DBSCAN(eps=0.1, minPts=4)
```

```python
clusters = scanner.fit_predict(data1.to_numpy())
```

```python
plt.scatter(data1.iloc[:, 0] , data1.iloc[:, 1] , c=clusters)
plt.show()
```
